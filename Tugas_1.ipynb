Import Libraries

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

Read Picture

```python
img_early = cv.imread('Bulba.jpg')
img_earlyRGB = cv.cvtColor(img_early,cv.COLOR_BGR2RGB)

height = img_early.shape[0]
width = img_early.shape[1]
channel = img_early.shape[2]
```

Show Picture that already converted RGB Since CV2 using BGR Format

```python
plt.imshow(img_earlyRGB)
plt.show()
```

Convert Picture to Grayscaled

```python
img_gray = cv.cvtColor(img_early,cv.COLOR_BGR2GRAY)
plt.imshow(img_gray,cmap="gray")
plt.show()
```

Setup counter Value for Histogram

```python
gray_counter = np.zeros(256,dtype=int)
equi_counter = np.zeros(256,dtype=int)
```

Gray

```python
for i in range(height):
  for j in range(width):
    gray_counter[img_gray[i][j]] += 1
```

```python
# plt.subplot(3,1,1)
plt.plot(gray_counter,'b',label="Gray Histogram")
plt.title("Gray Histogram")
plt.ylabel("Quantity")
plt.xlabel("Intensity")
plt.axis([0,255,0,gray_counter.max()])
plt.show()
```

Equalized Image

```python
image_equi = cv.equalizeHist(img_gray)
plt.imshow(image_equi,cmap="gray")
plt.show()
```

```python
for i in range (height):
  for j in range(width):
    equi_counter[image_equi[i][j]]+=1
```

```python
# plt.subplot(3,1,2)
plt.plot(equi_counter,'b',label="Equalized Histogram")
plt.title("Equalized Histogram")
plt.ylabel("Quantity")
plt.xlabel("Intensity")
plt.axis([0,255,0,equi_counter.max()])
plt.show()
```

Print Both Plot

```python
plt.subplot(3,1,1)
plt.plot(gray_counter,'b',label="Gray Histogram")
plt.ylabel("Quantity")
plt.xlabel("Intensity")
plt.axis([0,255,0,gray_counter.max()])

plt.subplot(3,1,2)
plt.plot(equi_counter,'b',label="Equalized Histogram")
plt.ylabel("Quantity")
plt.xlabel("Intensity")
plt.axis([0,255,0,equi_counter.max()])

plt.show()
```

Show Both Image Between Grayscaled and Equalized

```python
plt.imshow(img_gray,cmap="gray")
plt.title(label="Gray")
plt.show()
plt.imshow(image_equi,cmap="gray")
plt.title(label="Equalized")
plt.show()
```

Accumulation from All 

```python
plt.imshow(img_gray,cmap="gray")
plt.title(label="Gray")
plt.show()

plt.subplot(3,1,1)
plt.plot(gray_counter,'b',label="Gray Histogram")
plt.ylabel("Quantity")
plt.xlabel("Intensity")
plt.axis([0,255,0,gray_counter.max()])
plt.show()

plt.imshow(image_equi,cmap="gray")
plt.title(label="Equalized")
plt.show()

plt.subplot(3,1,2)
plt.plot(equi_counter,'b',label="Equalized Histogram")
plt.ylabel("Quantity")
plt.xlabel("Intensity")
plt.axis([0,255,0,equi_counter.max()])
plt.show()
```
